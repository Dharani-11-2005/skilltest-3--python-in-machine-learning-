```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

data = {
    'Strength':[75,50,90,40,85,70,30,60,80,20],
    'Frequency':[2000,1500,2200,1300,2100,1800,1200,1600,1900,1100],
    'Quality':['Good','Bad','Good','Bad','Good','Good','Bad','Bad','Good','Bad']
}

df = pd.DataFrame(data)

le = LabelEncoder()
df['Quality'] = le.fit_transform(df['Quality'])

plt.figure()
plt.scatter(df['Strength'], df['Frequency'])
plt.xlabel("Strength")
plt.ylabel("Frequency")
plt.title("Signal Quality Classification")
plt.show()
```
