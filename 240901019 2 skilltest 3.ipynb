```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

Voltage = [5,5.5,4.8,6,6.5,6.2,4.9,5.8,5.2,6.3]
Current = [0.5,0.4,0.6,0.3,0.2,0.25,0.55,0.35,0.5,0.22]
Status = [0,0,0,1,1,1,0,1,0,1]

X = np.array(list(zip(Voltage, Current)))
y = np.array(Status)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

plt.figure()
plt.scatter(Voltage, Current)
plt.scatter(5.4,0.45, marker='x')
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("Working vs Defective")
plt.show()
```
