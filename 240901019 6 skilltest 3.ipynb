```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([
    [4.8,0.8],[5.2,0.9],[4.7,0.95],[4.6,1.08],[4.8,0.87],
    [12.3,2.25],[12.36,2.38],[11.2,1.89],[12.17,1.82],[12.06,2.29]
])

model = KMeans(n_clusters=2)
model.fit(X)

labels = model.labels_

plt.figure()
plt.scatter(X[:,0], X[:,1])
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("K-Means Clustering")
plt.show()
```
